```python
norm = df.columns[df.columns != 'quality']
scaler = MinMaxScaler()
df[norm] = scaler.fit_transform(df[norm])
df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample LDA vector (Replace with your actual LDA vector)
lda_vector = np.array([0.7, 0.7])  # Example LDA vector

# Define data
X1 = np.array([[4, 1], [2, 4], [2, 3], [3, 6], [4, 4]])
X2 = np.array([[9, 10], [6, 8], [9, 5], [8, 7], [10, 8]])

# Project the data points onto the LDA vector
def project_data(data, lda_vector):
    return data @ lda_vector

X1_proj = project_data(X1, lda_vector)
X2_proj = project_data(X2, lda_vector)

# Plot
plt.figure(figsize=(12, 6))

# Original Data
plt.subplot(1, 2, 1)
plt.scatter(X1[:, 0], X1[:, 1], color='blue', label='X1')
plt.scatter(X2[:, 0], X2[:, 1], color='red', label='X2')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Original Data')
plt.legend()
plt.grid(True)
# Projected Data
plt.subplot(1, 2, 2)
plt.scatter(X1_proj, np.zeros_like(X1_proj), color='blue', label='X1 Proj')
plt.scatter(X2_proj, np.zeros_like(X2_proj), color='red', label='X2 Proj')
plt.axhline(0, color='black', linewidth=0.8)
plt.xlabel('Projection onto LDA Vector')
plt.title('Projected Data')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
plt.savefig('lda_projection2.png')
```
